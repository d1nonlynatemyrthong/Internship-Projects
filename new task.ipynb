```python
import pandas as pd
df=pd.read_csv("customer.csv")
print(df.to_string())
```

```python
print("Dataset dimension",df.shape,"\n\n")
print("\n\n")
df.info()
```

```python
df.columns=df.columns.str.strip().str.title().str.replace(' ','_')
print(df.columns)
```

```python
df.dropna(subset=["Name","Country","Email_Address","Joindate"],inplace=True)
df.drop_duplicates(subset=["Name","Country","Email_Address","Joindate"],inplace=True)
df.head(10)
```

```python
df.fillna({"Age":df["Age"].median(),"Income":df["Income"].mean()},inplace=True)
df["Age"]=df["Age"].astype(int)
df.loc[df["Income"]<30000,"Income"]=df["Income"].mean()
df.head(10)
```

```python
df["Gender"]=df["Gender"].str.strip().str.lower()
df["Country"]=df["Country"].str.strip().str.upper()
df["Name"]=df["Name"].str.strip().str.title()
df.head(10)
```

```python
df["Gender"].unique()
```

```python
g={"female":"F","male":"M","m":"M","f":"F"}
df["Gender"]=df["Gender"].map(g)
print(df["Gender"])
```

```python
df["Joindate"]=pd.to_datetime(df["Joindate"],format="mixed")
df.head(10)
```

```python
df.to_csv("customer_cleaned.csv", index=False)
```
